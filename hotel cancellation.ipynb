```python
import speech_recognition as sr
import pyttsx3
import time
from commands import get_time, search_wikipedia, tell_joke, open_app

# -------------------------
# TEXT-TO-SPEECH ENGINE
# -------------------------
engine = pyttsx3.init()
engine.setProperty("rate", 180)

def speak(text):
    print("Assistant:", text)
    engine.say(text)
    engine.runAndWait()

# -------------------------
# SPEECH RECOGNIZER
# -------------------------
recognizer = sr.Recognizer()

def listen():
    with sr.Microphone() as source:
        print("Listening...")
        recognizer.adjust_for_ambient_noise(source, duration=0.3)
        audio = recognizer.listen(source)

    try:
        text = recognizer.recognize_google(audio)
        print("You:", text)
        return text.lower()
    except:
        return ""

# -------------------------
# COMMAND HANDLER
# -------------------------
def handle_command(cmd):
    if "time" in cmd:
        speak(get_time())

    elif "search" in cmd:
        query = cmd.replace("search", "").strip()
        speak("Searching Wikipedia...")
        speak(search_wikipedia(query))

    elif "joke" in cmd:
        speak(tell_joke())

    elif "open" in cmd:
        app = cmd.replace("open", "").strip()
        speak(open_app(app))

    elif "stop" in cmd or "goodbye" in cmd:
        speak("Goodbye!")
        exit()

    else:
        speak("I didn't understand. Could you repeat?")

# -------------------------
# MAIN LOOP (Wake Word)
# -------------------------
def main():
    speak("Hello! Say 'assistant' to wake me up.")

    while True:
        text = listen()

        if "assistant" in text:
            speak("I'm listening.")
            command = listen()
            handle_command(command)

        time.sleep(0.2)

if _name_ == "_main_":
    main()
```
